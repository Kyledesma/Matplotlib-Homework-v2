## Observations and Insights 

```python
#Observation 1: mice treated with capomulin had the second lowest average mean and median tumor volume
#Observation 2: Infubinol had one outlier with a significantly lower end tumor volume. Would be worth investigating what was different with this mouse that led to such low results, can build more studies off of this
#Observation 3: There was a moderately strong (r^2 = .7) positive relationship between what the mouse's weight was and what the end tumor volume size; Potentially shows that being heavier can lead to more SCC's
```

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
micedata_df = pd.merge(mouse_metadata,study_results, on = 'Mouse ID')
micedata_df
```

```python
# Checking the number of mice.
micedata_df["Mouse ID"].count()
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mouse_ids = micedata_df.loc[micedata_df.duplicated(subset=['Mouse ID', 'Timepoint']),'Mouse ID'].unique()
duplicate_mouse_ids[0]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_mousedata_df = micedata_df[micedata_df['Mouse ID'].isin(duplicate_mouse_ids)==False]
```

```python
# Checking the number of mice in the clean DataFrame.
clean_mousedata_df["Mouse ID"].count()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
regimen_df = clean_mousedata_df.groupby("Drug Regimen")

#Calculating Mean
mean = regimen_df["Tumor Volume (mm3)"].mean()


#Calculating Median
median = regimen_df["Tumor Volume (mm3)"].median()


# Calculating Variance
var = regimen_df["Tumor Volume (mm3)"].var()


#Calculating Standard Deviance
stdev = regimen_df["Tumor Volume (mm3)"].std()


#Calculating SEM
sem = regimen_df["Tumor Volume (mm3)"].sem()


# This method is the most straightforward, creating multiple series and putting them all together at the end.
summary_df = pd.DataFrame([mean,median,var,stdev,sem],
                         index = ['Mean','Median','Variance','Standard Deviation','Std Error of Mean'])

summary_df = summary_df.T

summary_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
round (regimen_df['Tumor Volume (mm3)'].agg(['mean','median','var','std','sem']), 2)
# This method produces everything in a single groupby function
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
regimens_df = clean_mousedata_df.loc[clean_mousedata_df["Drug Regimen"].isin(["Capomulin", "Ramicane", "Infubinol", "Ceftamin"])]
mouse_group = regimens_df.groupby("Mouse ID")



# Start by getting the last (greatest) timepoint for each mouse
last_timepoint = mouse_group["Timepoint"].max()



# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
combined_df = pd.merge(last_timepoint, regimens_df, on = 'Mouse ID')
final_timepoint_df = combined_df.loc[combined_df["Timepoint_x"] == combined_df["Timepoint_y"]]

final_timepoint_df
```

```python
# Create empty list to fill with tumor vol data (for plotting)
tumor = []
mice = []
regimen = []

tumor = final_timepoint_df["Tumor Volume (mm3)"]
mice = final_timepoint_df["Mouse ID"]
regimen = final_timepoint_df["Drug Regimen"]


#Creating a new DF of final volume for mice in the 4 specified Drug regimens
boxplot_df = pd.DataFrame({"Mouse ID" : mice, "Final Tumor Volume" : tumor, "Drug Regimen" : regimen})

#Extracting the 4 different drug regimens 
capomulin_df = boxplot_df.loc[boxplot_df["Drug Regimen"] == "Capomulin"]
ramicane_df = boxplot_df.loc[boxplot_df["Drug Regimen"] == "Ramicane"]
infubinol_df = boxplot_df.loc[boxplot_df["Drug Regimen"] == "Infubinol"]
ceftamin_df = boxplot_df.loc[boxplot_df["Drug Regimen"] == "Ceftamin"]



# Plotting the 4 box plots
cap_volume = capomulin_df["Final Tumor Volume"]
ram_volume = ramicane_df["Final Tumor Volume"]
infub_volume = infubinol_df["Final Tumor Volume"]
ceft_volume = ceftamin_df["Final Tumor Volume"]

data_to_plot = [cap_volume, ram_volume, infub_volume, ceft_volume]

fig1, ax1 = plt.subplots()
ax1.set_title('Final Tumor Volume across Drugs')
ax1.set_ylabel('Final Tumor Volume (mm3)')
ax1.boxplot(data_to_plot)
plt.show()

```

```python
#Capomulin outlier calculation
quartiles = cap_volume.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)

cap_outliers = capomulin_df.loc[(capomulin_df["Final Tumor Volume"] > upper_bound) | (capomulin_df["Final Tumor Volume"] < lower_bound)]


#Ramicane outlier calculation
quartiles = ram_volume.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)

ram_outliers =  ramicane_df.loc[(ramicane_df["Final Tumor Volume"] < lower_bound) | (ramicane_df["Final Tumor Volume"] > upper_bound)]


#Infubinol outlier calculation
quartiles = infub_volume.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)

infub_outliers =  infubinol_df.loc[(infubinol_df["Final Tumor Volume"] < lower_bound) | (infubinol_df["Final Tumor Volume"] > upper_bound)]


#Ceftamin outlier calculation
quartiles = ceft_volume.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)

ceft_outliers =  ceftamin_df.loc[(ceftamin_df["Final Tumor Volume"] < lower_bound) | (ceftamin_df["Final Tumor Volume"] > upper_bound)]




#Combining and printing all outliers
outliers = pd.concat([cap_outliers,ram_outliers,infub_outliers,ceft_outliers])
print(outliers)
```

## Line and Scatter Plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
capomulin_df = clean_mousedata_df.loc[clean_mousedata_df["Drug Regimen"] == 'Capomulin']

#Grouping by Mouse ID and getting a list of all unique Mice, then picking the first one
cap_group = capomulin_df.groupby("Mouse ID")
cap_unique = cap_group["Mouse ID"].unique()
cap_mouse = cap_unique[0]
cap_mouse = cap_mouse[0]

#Extracting the data
mouse_df = clean_mousedata_df.loc[clean_mousedata_df["Mouse ID"] == cap_mouse]


#Plotting the graph
x_values = mouse_df['Timepoint']
y_values = mouse_df['Tumor Volume (mm3)']
plt.plot(x_values,y_values)

#Labeling graph
plt.xlabel("Timepoint")
plt.ylabel("Tumor Volume (mm3)")
plt.title(f"Data for Mouse {cap_mouse}")
plt.legend(['Tumor Volume'])
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
capomulin_df = clean_mousedata_df.loc[clean_mousedata_df["Drug Regimen"] == 'Capomulin']
mice_df = capomulin_df.groupby("Mouse ID")

mice_df.head()

mouse_weight = mice_df['Weight (g)'].mean()
mean_tumor = mice_df['Tumor Volume (mm3)'].mean()





#Creating a new dataframe
weight_df = pd.DataFrame({"Mouse Weight": mouse_weight, "Average Tumor Volume": mean_tumor})


#Creating x and y values
x_values = weight_df["Mouse Weight"]
y_values = weight_df["Average Tumor Volume"]


#Plotting
plt.scatter(x_values,y_values)
plt.xlabel("Mouse Weight (g)")
plt.ylabel("Average Tumor Volume")
plt.title("Relationship between Mouse Weight and Average Tumor Volume")
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)

#Get regression values
regress_values = x_values * slope + intercept

#Create line equation string
line_eq = "y = " + str(round(slope,2)) + "x +" + str(round(intercept,2))

#Create updated plot
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")


#Formatting
plt.xlabel("Mouse Weight (g)")
plt.ylabel("Average Tumor Volume")
plt.title("Relationship between Mouse Weight and Average Tumor Volume")
plt.annotate(line_eq,(20,15),fontsize=15,color="red")

# Print r square value
print(f"The r-squared is: {rvalue**2}")

plt.show()
```
